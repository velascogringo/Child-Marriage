```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#import libraries
```

```python
file = '/content/drive/MyDrive/Colab Notebooks/viz5_child_marriage.csv'
df = pd.read_csv(file)
df.head() #show 1st five
```

```python
df.shape #128 rows, 4 columns
```

```python
top10fem15 = df.sort_values(['females_married_by_15'], ascending= False).head(10)
top10fem15[['country','females_married_by_15']]
#look for top 10 countries where females are getting married by age of 15
```

```python
fig, ax = plt.subplots(figsize = (15,7))
sns.barplot(data=top10fem15, x='country', y='females_married_by_15')
ax.set_xlabel('Country')
ax.set_ylabel('Percentage')
ax.set_title('Top 10 Countries with Females Getting Married by Age of 15')
ax.set_xticklabels(ax.get_xticklabels(),rotation = 30)
plt.show();

#create bar chart of top 10 fem getting married by 15
```
